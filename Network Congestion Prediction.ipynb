## Our dataset lack some core Quality of Service (QoS) feature metrics like latency, packet loss, and throughput degradation. To train a model to predict network degradation and make optimization decisions, those specific target variables must exist in the dataset. Therefore we will have to engineer it out ourself with some mathematical features.

### Aggregating Total Network Load
To model the physical stress on each cell tower, we first consolidate the disparate traffic features (SMS, Voice, and Internet) into a unified throughput metric. In telecommunications, hardware queues process total packet volume regardless of the payload type, making this aggregate our primary independent variable.

```python
df = df.groupby(['CellID', 'datetime']).agg({
    'smsin': 'sum',
    'smsout': 'sum',
    'callin': 'sum',
    'callout': 'sum',
    'internet': 'sum'
}).reset_index()
```

To ensure temporal consistency in time-series feature extraction, the dataset was sorted by CellID and timestamp prior to applying rolling window operations. This guarantees that computed features such as expected throughput accurately reflect historical network behavior.”

```python
df = df.sort_values(by=['CellID', 'datetime'])
```

```python
# T = Sin + Sout + Cin + Cout + I
# T = throughput
# I = internet
# Throughput represents the total network load

df['throughput'] = (
    df['smsin'] + df['smsout'] +
    df['callin'] + df['callout'] +
    df['internet']
)
```

```python
df.head()
```

## Latency Simulation Logic
This section simulates network Latency based on observed Throughput levels for each cell. Since real-world latency typically increases as a network becomes more congested, we use a two-step mathematical approach:
## Traffic Normalization (traffic_norm):
We apply Min-Max Scaling to the throughput, grouped by CellID. This converts raw data into a utilization index between 0 and 1.

0 represents the cell at its lowest load (idle).

1 represents the cell at its maximum recorded capacity (peak traffic).

Note: A small epsilon (1e^-9) is added to prevent division-by-zero errors.

## Simulating Latency using M/M/1 Queuing Theory
## Router buffer latency is rarely linear. According to network queuing theory, latency remains stable under normal loads but spikes exponentially as the hardware approaches maximum capacity. We model this physical behavior using a cubic function ($x^3$), establishing a base ping of 50ms that non-linearly scales to ~250ms under extreme network stress.


```python
# Normalization per cell
df['traffic_norm'] = df.groupby('CellID')['throughput'].transform(
    lambda x: (x - x.min()) / (x.max() - x.min() + 1e-9)
)
```

## Linear Latency Modeling:
We estimate latency using the heuristic formula: Latency = 50 +(Traffic_Norm x 200).

Base Latency (50ms): Represents the minimum hardware/propagation delay in an uncongested state.

Congestion Penalty (200ms): Simulates the additional queuing delay experienced as the cell reaches 100% utilization.

Result: This creates a realistic academic proxy where latency scales linearly from 50ms (idle) to 250ms (saturated).

```python
# Non-linear latency
df['latency_ms'] = 50 + 200 * (df['traffic_norm'] ** 3)
```

## Network Volatility (Traffic Change)

This metric identifies sudden spikes or drops in network demand by calculating the first-order difference in throughput.

## Modeling Packet Loss via Buffer Overflows and Traffic Bursts
#### Packet loss occurs under two conditions: sustained maximum capacity or sudden, intense traffic spikes that overwhelm the router before it can adjust. This hybrid feature calculates loss by checking a hard physical threshold (85% capacity) and factoring in the intensity of sudden traffic bursts (calculated via the first discrete difference, diff(), of the throughput).

```python
# df['traffic_diff'] = df.groupby('CellID')['throughput'].diff().fillna(0)
# burst_intensity = abs(df['traffic_diff']) / (df['throughput'].max() + 1e-9)

# Traffic difference
df['traffic_diff'] = df.groupby('CellID')['throughput'].diff().fillna(0)

# Burst intensity (fixed)
df['burst_intensity'] = df.groupby('CellID')['traffic_diff'].transform(
    lambda x: abs(x) / (abs(x).max() + 1e-9)
)
```

```python
# Packet loss
df['packet_loss_pct'] = np.where(
    df['traffic_norm'] > 0.70,
    (df['traffic_norm'] * 0.4) + (df['burst_intensity'] * 0.6),
    0.0
)

df['packet_loss_pct'] = df['packet_loss_pct'].clip(0, 1)
```

## Contextual Anomaly Detection (Rolling Averages)
#### A network might be degraded even if it hasn't completely crashed. To teach the AI to recognize subtle performance dips, we calculate a 5-step rolling mean to establish the "expected" throughput for that specific time window. The delta between the expected throughput and actual throughput gives us a quantifiable degradation metric.

```python
# df['expected_throughput'] = df.groupby('CellID')['throughput'].transform(
#     lambda x: x.rolling(window=5, min_periods=1).mean()
# )
# df['throughput_degradation'] = (df['expected_throughput'] - df['throughput']).clip(lower=0)
```

```python
# Throughput expectation
df['expected_throughput'] = df.groupby('CellID')['throughput'].transform(
    lambda x: x.rolling(window=5, min_periods=1).mean()
)
```

```python
# Degradation
df['throughput_degradation'] = (
    df['expected_throughput'] - df['throughput']
).clip(lower=0)
```

## Defining the ML Target Variable ($Y$)
#### This boolean feature serves as the ground truth label for our predictive AI model. We classify the network state as mathematically "Congested" (1) if it meets any of our strict Quality of Service (QoS) failure conditions: severe latency (> 150ms), tangible packet loss (> 0.1%), or significant throughput degradation.

```python
# Congestion target
df['congestion'] = (
    (df['latency_ms'] > 150) |
    (df['packet_loss_pct'] > 0.1) |
    (df['throughput_degradation'] > df['throughput_degradation'].quantile(0.85))
).astype(int)
```
